```python
import nltk
nltk.download('punkt')
nltk.download('vader_lexicon')
#from textblob import TextBlob
import pandas as pd
import numpy as np
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import subjectivity
from nltk.sentiment import SentimentAnalyzer
from nltk.sentiment.util import *
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import tokenize

sentiment_analysis_columns = ['negativity', 'neutrality', 'positivity', 'compound']

file_names = [...]

sid = SentimentIntensityAnalyzer()

for i in range(len(file_names)):
    expanded_text_dataset = []
    file_name = file_names[i]
    print('Sentiment analysis for: %s' %file_name)
    file_df = reviews_df
    column_names = list(file_df.columns.values)
    all_columns = column_names + sentiment_analysis_columns
    text_segment_list = list(file_df['comments'])
    for j in range(len(text_segment_list)):
        row_info = list(file_df.iloc[j])
        text = text_segment_list[j]
        ##Sentiment analysis
        try:
            ss = sid.polarity_scores(text)
            negativity = ss['neg']
            neutrality = ss['neu']
            positivity = ss['pos']
            compound = ss['compound']
            temp_data = row_info + [negativity, neutrality, positivity, compound]
            expanded_text_dataset.append(temp_data)
        except:
            print('Sentiment analysis not done: ' + str(i) + ' / ' + str(len(file_df)))
            temp_data = row_info + ['NA', 'NA', 'NA', 'NA']
            expanded_text_dataset.append(temp_data)
    ##Storing the results in a dataset
    sentiment_dataset_df = pd.DataFrame(expanded_text_dataset, columns = all_columns)
    #sentiment_dataset_df.to_excel('%s_sentiment_analysis.xlsx' %file_name)
```

```python
#Q5
# We firstly create the first candidate, 1 item for each candidate, and use frozensets to create a list for each unique item. Then, find the support counts of c1.
def create_c1(transactions):
    c1 = []
    c1_counts = {}
    for transaction in transactions:
        for item in transaction:
            item = frozenset([item]) # for each item in transaction, make it to a set
            if item not in c1:
                c1.append(item) # if item is not in c1 before, we add it to c1
            if item in c1_counts:
                c1_counts[item] += 1 # if itemset is already in the item count, then we add it's item count by 1
            else:
                c1_counts[item] = 1 # if itemset is not in the item count, then it's the first time we count it, so we set item count of this item to 1

    return c1, c1_counts

# Create frequent itemsets that pass the minSup
def create_freq_itemset(transactions, ck, minSup):
    itemset_count = {}
    nrows = len(transactions)  
# Go through the transaction and count the occurance of each item
    for transaction in transactions:
        for item in ck:
            if item.issubset(transaction):
                if item in itemset_count: 
                    itemset_count[item] += 1 # if itemset is already in the item count, then we add it's item count by 1
                else: 
                    itemset_count[item] = 1 # if itemset is not in the item count, then it's the first time we count it, so we set item count of this item to 1
   
    freq_itemset = []
    itemset_support = {}
                  
    for item in itemset_count:
        support = itemset_count[item] / nrows # calculate support of each item
        if support >= minSup:
            freq_itemset.append(item) # if the support of this item is greater than minSup, then it is a frequent item
            itemset_support[item] = support # for the frequent items, store their supports
    return freq_itemset, itemset_support

def create_ck(freq_k_transaction, k_next):
    ck = []
    k=len(freq_k_transaction)
    freq_set = [list(i) for i in freq_k_transaction]

    for i in range(0,k-1):
        for j in range(i + 1, k):
            for a in freq_set[j]:
                new_set = freq_set[i] + [a]
                if len(new_set) == k_next:
                    ck.append(new_set)
    return ck

def apriori(transactions, minSup, minConf):
    
    c1, c1_counts = create_c1(transactions) # create c1 and c1 counts
    freq_item, item_support_dict = create_freq_itemset(transactions, c1, minSup) # find frequent itemsets for c1 and their supports
    freq_items = [freq_item]
    
    k = 0
    while len(freq_items[k]) > 0:
        freq_item = freq_items[k]
        ck = create_ck(freq_item, k) #create ck      
        freq_item, item_support = create_freq_itemset(transactions, ck, minSup) # find frequent itemsets for ck and their supports
        freq_items.append(freq_item) # add frequent itemsets for ck to freq_item[]
        item_support_dict.update(item_support) # add their supports to item_support{}
        k = k + 1
        
    return freq_items, item_support_dict
```

```python
#Q6
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
listings_df1=listings_df.dropna(subset=['host_response_rate','review_scores_rating','review_scores_accuracy','review_scores_cleanliness','review_scores_checkin','review_scores_communication','positivity_mean','negativity_mean','positivity_simple_mean', 'negativity_simple_mean'])
y=list(listings_df1['price'])
x1=list(listings_df1['host_response_rate'])
x2=list(listings_df1['review_scores_rating'])
x3=list(listings_df1['review_scores_accuracy'])
x4=list(listings_df1['review_scores_cleanliness'])
x5=list(listings_df1['review_scores_checkin'])
x6=list(listings_df1['review_scores_communication'])
x7=list(listings_df1['positivity_mean'])
x8=list(listings_df1['negativity_mean'])
x9=list(listings_df1['positivity_simple_mean'])
x10=list(listings_df1['negativity_simple_mean'])
X=np.column_stack((x1,x2,x3,x4,x5,x6,x7,x8,x9,x10))
X=sm.add_constant(X)
results = sm.OLS(y,X).fit()
print(results.summary())
```

```python
#The R-square is 0.05. According to to P-value in the chart and 5 percent confidence interval, we find the P- value of the coefficients of constant, review_scores_rating, review_scores_accuracy, review_scores_cleanliness, review_scores_checkin, positivity_simple_mean, and negativity_simple_mean are less than the the significance level 0.05. Thus, these coefficients are significant. This means the price y can be closely related to these variables.
#Effect size is a number measuring the strength of the relationship between two variables in a population, or a sample-based estimate of that quantity.
```

```python
#Q7
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

pca = PCA(n_components = 3)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
explained_variance = pca.explained_variance_ratio_
X_train=sm.add_constant(X_train)
pca_results = sm.OLS(y_train,X_train).fit()
print(pca_results.summary())
```

```python
#As doing PCA, the standard errors of coefficients are generally smaller than before, which means the predictors are performs better in the model. 
#However, the R-square reduces a lot from 0.05 to 0.001, which means this model does not do well in predicting the price.
```
